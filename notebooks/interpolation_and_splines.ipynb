# <span style="color:rgb(213,80,0)">Solution to question 1</span>

```matlab
fprintf("Solution to Question 1\n");


x = [0.5, 1, 1.5, 2];
y = cos(log(x));
target = 1.75;
true_val = cos(log(target));


% Date Points for Plotting
x_points = linspace(0.5, 2, 500);
f_true = cos(log(x_points));


% degree atmost 1
x1 = [1.5, 2];
y1 = cos(log(x1));
p1 = lagrange_interpolation(x1, y1, target);
err1 = abs(true_val - p1);
p1_vals = zeros(size(x_points));
for k = 1:length(x_points)
    p1_vals(k) = lagrange_interpolation(x1, y1, x_points(k));
end


% degree atmost 2
x2 = [1.0, 1.5, 2];
y2 = cos(log(x2));
p2 = lagrange_interpolation(x2, y2, target);
err2 = abs(true_val - p2);
p2_vals = zeros(size(x_points));
for k = 1:length(x_points)
    p2_vals(k) = lagrange_interpolation(x2, y2, x_points(k));
end


% degree atmost 3
x3 = [0.5, 1.0, 1.5, 2];
y3 = cos(log(x3));
p3 = lagrange_interpolation(x3, y3, target);
err3 = abs(true_val - p3);
p3_vals = zeros(size(x_points));
for k = 1:length(x_points)
    p3_vals(k) = lagrange_interpolation(x3, y3, x_points(k));
end


fprintf("Value of f(x) at 1.75: %.10f\n", true_val);
fprintf("Value by polynomial of degree 1: %.10f, Error: %.10e\n", p1, err1);
fprintf("Value by polynomial of degree 2: %.10f, Error: %.10e\n", p2, err2);
fprintf("Value by polynomial of degree 3: %.10f, Error: %.10e\n", p3, err3);


% Plotting Code
% Not using black because it is not visible against the black background
plot(x_points, f_true, 'LineWidth', 2);
hold on;
plot(x_points, p1_vals, '--', 'LineWidth', 1.5);
plot(x_points, p2_vals, '--', 'LineWidth', 1.5);
plot(x_points, p3_vals, '--', 'LineWidth', 1.5);
plot(x, y, 'ro', 'MarkerSize', 8, 'MarkerFaceColor', 'r');
hold off;
legend('f(x) = cos(ln(x))', 'degree 1', 'degree 2', 'degree 3', 'Data Points');
xlabel('x');
ylabel('y');
title('Lagrange Interpolation');
grid on;




function val = lagrange_interpolation(x_nodes, y_nodes, x)
    n = length(x_nodes);
    val = 0;
    for i = 1:n
        L = 1;
        for j = 1:n
            if j ~= i
                L = L * ((x - x_nodes(j))/(x_nodes(i) - x_nodes(j)));
            end
        end
        val = val + L*y_nodes(i);
    end
end
```


## Solution to Question 2

```matlab
fprintf("\nSolution to Question 2\n");
x = [-0.10, 0.00, 0.20, 0.30];
f = [17.3000, 2.0000, 5.1900, 1.0000];


coef = divided_diff(x, f);
f1 = newton(coef, x, 0.1);
f2 = newton(coef, x, 0.4);


fprintf("f(0.1) = %.10f\n", f1);
fprintf("f(0.4) = %.10f\n", f2);


% Plotting code
x_points = linspace(-0.2, 0.5, 500);
y_points = zeros(size(x_points));
for i = 1:length(x_points)
    y_points(i) = newton(coef, x, x_points(i));
end
figure;
plot(x_points, y_points, 'LineWidth', 2);
hold on;
plot(x, f, 'ro', 'MarkerSize', 8, 'MarkerFaceColor', 'r');
legend('Interpolating Poly.', 'Data Points');
xlabel('x');
ylabel('y');
title('Newton Interpolation');
hold off;
grid on;


% Part d
xd = [-0.10, 0.00, 0.05, 0.20, 0.30];
fd = [17.3000, 2.0000, 3.1250, 5.1900, 1.0000];
coefd = divided_diff(xd, fd);
f1d = newton(coefd, xd, 0.1);
f2d = newton(coefd, xd, 0.4);


fprintf("\nf(0.1) for part d = %.10f\n", f1d);
fprintf("f(0.4) for part d = %.10f\n", f2d);


% Plotting code for part d
x_pointsd = linspace(-0.2, 0.5, 500);
y_pointsd = zeros(size(x_points));
for i = 1:length(x_pointsd)
    y_pointsd(i) = newton(coefd, xd, x_pointsd(i));
end
figure;
plot(x_pointsd, y_pointsd, 'LineWidth', 2);
hold on;
plot(xd, fd, 'ro', 'MarkerSize', 8, 'MarkerFaceColor', 'r');
legend('Interpolating Poly. (d)', 'Data Points');
xlabel('x');
ylabel('y');
title('Newton Interpolation (d)');
hold off;
grid on;


function coef = divided_diff(x, y)
    n = length(x);
    table = zeros(n, n+1);
    table(:, 1) = x(:);
    table(:, 2) = y(:);


    for j = 3:n+1
        for i = 1:n-j+2
            table(i, j) = (table(i+1, j-1) - table(i, j-1))/(table(i+j-2, 1) - table(i, 1));
        end
    end
    coef = table(1, 2:end);


end


function val = newton(coef, x_nodes, x)
    n = length(coef);
    val = coef(1);
    prod = 1;
    for i = 2:n
        prod = prod*(x - x_nodes(i-1));
        val = val + coef(i)*prod;

    end


end
```


## Solution to Question 3

```matlab
fprintf("\nSolution to Question 3\n")


a = [0, 23.04, 47.37, 97.45, 123.66];
x = [0, 0.25, 0.5, 1.00, 1.25];
n = length(a);
h = diff(x);
alpha = zeros(1, n);
for i = 2:n-1
    alpha(i) = (3*(a(i+1) - a(i))/h(i)) - (3*(a(i) - a(i-1))/h(i-1));
end
l = zeros(1, n);
m = zeros(1, n);
z = zeros(1, n);
l(1) = 1;
m(1) = 0;
z(1) = 0;


for i = 2:n-1
    l(i) = 2*(x(i+1) - x(i-1)) - h(i-1)*m(i-1);
    m(i) = h(i)/l(i);
    z(i) = (alpha(i) - h(i-1)*z(i-1))/l(i);
end
l(n) = 1;
z(n) = 0;


b = zeros(1, n-1);
c = zeros(1, n);
d = zeros(1, n-1);
c(n) = 0;


for j = n-1:-1:1
    c(j) = z(j) - m(j)*c(j+1);
    b(j) = (a(j+1) - a(j))/h(j) - h(j)*(c(j+1) + 2*c(j))/3;
    d(j)= (c(j+1) - c(j))/(3*h(j));
end


% part b
xp = 0.75;
j = 3;
S = a(j) + b(j)*(xp - x(j)) + c(j)*(xp - x(j))^2 + d(j)*(xp - x(j))^3;


fprintf("Predicted time at 0.75 miles: %.4f seconds\n", S);
fprintf("Actual time: 71.80 seconds\n");
fprintf("Error: %.4f seconds\n", abs(S - 71.80));


% Part c
x_start = 0;
j = 1;
v_start = b(j) + 2*c(j)*(x_start - x(j)) + 3*d(j)*(x_start - x(j))^2;
fprintf("Starting speed: %.4f sec/mile\n", v_start);
x_finish = 1.25;
j = 4;
v_end = b(j) + 2*c(j)*(x_finish - x(j)) + 3*d(j)*(x_finish - x(j))^2;
fprintf("Finishing speed: %.4f sec/mile\n", v_end);
```
