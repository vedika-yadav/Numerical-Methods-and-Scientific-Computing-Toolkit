```matlab
% code to call plotting function
plot_error_hilb(3, 9)


% function to calculate error and condition number
function [e,k] = Error_hilb(n)
A = hilb(n);
xtrue = ones(n,1);
b = A * xtrue;
xcomp = A\b;
e = norm(xcomp - xtrue, inf);
k = cond(A);
end


%function to plot loglog plot of err and normal plot of condition number
function [] = plot_error_hilb(n1, n2)
N = n1:n2;
err = zeros(size(N));
condA = zeros(size(N));


for i = 1:length(N)
    n = N(i);
    [err(i), condA(i)] = Error_hilb(n);
end


figure
loglog(N, err, 'o-')
xlabel('Matrix Size')
ylabel('Error (Infinity Norm)')
title('Error vs Size of Hilbert Matrices')
grid on


figure
plot(N, condA, 'o-')
xlabel('Matrix Size')
ylabel('Condition Number')
title('Condition Number vs Size of Hilbert Matrix')
grid on


end
```

```matlab
plot_cond_no_vand(5, 25)


function k = cond_no_vand(n)
x = (0:n)'/n;
V = zeros(n+1, n+1);
for j = 1:n+1
    V(:, j) = x.^(j-1);
end
k = cond(V);
end


function [] = plot_cond_no_vand(n1, n2)
N = n1:n2;
condV = zeros(size(N));
for i = 1:length(N)
    n = N(i);
    condV(i) = cond_no_vand(n);
end


figure
plot(N, condV, 'o-')
xlabel('n')
ylabel('Condition Number')
title('Condition Number vs Matrix Size (Vandermonde)')
grid on
end


% Part b


plot_cond_no_tridiag(5, 100)


function k = cond_no_tridiag(n)
T = zeros(n, n);
for j = 2:n-1
    T(j, j-1) = -1;
    T(j, j) = 2;
    T(j, j + 1) = -1;
end
T(1, 1) = 2;
T(1, 2) = -1;
T(n, n) = 2;
T(n, n-1) = -1;
k = cond(T);
end


function [] = plot_cond_no_tridiag(n1, n2)
N = n1:n2;
condT = zeros(size(N));
for i = 1:length(N)
    n = N(i);
    condT(i) = cond_no_tridiag(n);
end


figure
plot(N, condT, 'o-')
xlabel('Matrix Size')
ylabel('Condition Number')
title('Condition Number vs Matrix Size (Tridiagonal Matrix)')
grid on
end
```
