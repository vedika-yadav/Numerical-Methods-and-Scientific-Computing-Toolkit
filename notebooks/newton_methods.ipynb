# <span style="color:rgb(213,80,0)">Question 1</span>

```matlab
fprintf("Solution to Question 1\n");
f = @(x) (x^2)*abs(sin(x)) - 4;


x_pts = linspace(0, 4, 200);
y = (x_pts.^2).*abs(sin(x_pts)) - 4;
plot(x_pts, y)
title("Graph of function")
xlabel("x")
ylabel("f")
grid on


p0 = 3.6;
p1 = 3.7;
secant_method(p0, p1);
p0 = 2.8;
p1 = 2.9;
secant_method(p0, p1);


function [] = secant_method(p0, p1)
    f = @(x) (x^2)*abs(sin(x)) - 4;
    TOL = 10^(-6);
    N0 = 60;
    q0 = f(p0);
    q1 = f(p1);
    for i = 2:N0
        p = p1 - q1*(p1 - p0)/(q1 - q0);
        if abs(p - p1) < TOL
            fprintf("Tolerance: %e; Approximation: %.6f, Iterations: %d\n", TOL, p, i);
            break;
        end
        p0 = p1;
        q0 = q1;
        p1 = p;
        q1 = f(p1);
    end

    if(i == N0)
        fprintf("The method failed to converge after %d iterations.\n", N0);
    end
end
```


## Question 2

```matlab
fprintf("\nSolution to Question 2\n");
syms x;
f_sym = x^2 - 2*(exp(-x))*x + exp(-2*x);
g_sym = diff(f_sym);


f = matlabFunction(f_sym);
g = matlabFunction(g_sym);


p0 = 1;
TOL = 10^(-8);
N0 = 30;
for i = 1:N0
    p = p0 - f(p0)/g(p0);
    if (abs(p - p0) < TOL)
        fprintf("Tolerance: %e; Approximation: %.8f, Iterations: %d\n", TOL, p, i);
        break;
    end
    p0 = p;
end
if i == N0
    fprintf("The method failed to converge after %d iterations\n", N0);
end
```


## Question 3

```matlab
fprintf("\nSolution to Question 3\n");
syms x1 x2 x3;
f1_sym = 6*x1 - 2*cos(x2*x3) - 1;
f2_sym = 9*x2 + sqrt(x1^2 + sin(x3) + 1.06) + 0.9;
f3_sym = 60*x3 + 3*exp(-x1*x2) + 10*pi - 3;


F_sym = [f1_sym; f2_sym; f3_sym];
J_sym = jacobian(F_sym, [x1, x2, x3]);


F = matlabFunction(F_sym, 'Vars', {[x1; x2; x3]});
J = matlabFunction(J_sym, 'Vars', {[x1; x2; x3]});


x = [0;0;0];
TOL = 10^(-6);
N0 = 50;
for i = 1:N0
    F_val = F(x);
    J_val = J(x);
    s = -J_val\F_val;
    x = x + s;
    if norm(s, inf) < TOL
        fprintf("Tolerance: %e; Approximation: [%.6f, %.6f, %.6f], Iterations: %d\n", TOL, x(1), x(2), x(3), i);
        break;
    end

end
if(i == N0)
    fprintf("The method failed to converge after %d iterations\n", N0);
end
```
