# <span style="color:rgb(213,80,0)">Solution to question 1</span>

```matlab
fprintf('Solution to Question 1\n');
A = [10, -1, 2, 0; -1, 11, -1, 3; 2, -1, 10, -1; 0, 3, -1, 8];
b = [6; 25; -11; 15];
x = [0; 0; 0; 0];
x1 = zeros(size(x));
TOL = 10^(-4);
for k = 1:30
    for i = 1:4
        sum_here = 0;
        for j = 1:4
            if(j ~= i)
                sum_here = sum_here + A(i, j)*x(j);
            end
        end
        x1(i) = (b(i) - sum_here)/A(i, i);
    end
    if (norm(b - A*x1, 2)/ norm(b, 2)) < TOL
        fprintf("The answer vector is [%.4f, %.4f, %.4f, %.4f] in %d iterations\n", x1(1), x1(2), x1(3), x1(4), k);
        break;
    end
    x = x1;
end
if k == 30
    fprintf("Result not found\n");
end


% Gauss Seidal
x = [0; 0; 0; 0];
x2 = [0;0;0;0];


for k = 1:30
    for i = 1:4
        sum1 = 0;
        sum2 = 0;
        for j = 1:(i-1)
            sum1 = sum1 + A(i, j)*x2(j);
        end
        for j = (i+1):4
            sum2 = sum2 + A(i, j)*x(j);
        end


        x2(i) = (b(i) - sum1 - sum2)/A(i, i);
    end
    if (norm(b - A*x2, 2)/ norm(b, 2)) < TOL
        fprintf("The answer vector is [%.4f, %.4f, %.4f, %.4f] in %d iterations\n", x2(1), x2(2), x2(3), x2(4), k);
        break;
    end
    x = x2;
end
if k == 30
    fprintf("Result not found\n");
end


D = diag(diag(A));
L = -tril(A, -1);
U = -triu(A, 1);


Mj = D\(L + U);
Mgs = (D - L)\U;
rhoJ = max(abs(eig(Mj)));
rhoGS = max(abs(eig(Mgs)));


fprintf("The spectral radii are %.6f and %.6f\n", rhoJ, rhoGS);
```


## Solution to Question 2

```matlab
fprintf('\nSolution to Question 2\n');
A = [4,-1,0,0,0,0; -1,4,-1,0,0,0; 0,-1,4,0,0,0; 0,0,0,4,-1,0; 0,0,0,-1,4,-1; 0,0,0,0,-1,4];
b = [0;5;0;6;-2;6];


x = [0;0;0;0;0;0];
x1 = x;
TOL = 10^(-8);
D = diag(diag(A));
L = -tril(A, -1);
U = -triu(A, 1);
Mj = D\(L+U);
rhoJ = max(abs(eig(Mj)));
w5 = 2/(1+sqrt(1 - (rhoJ^2)));
w = [1.01, 1.05, 1.10, 1.15, w5];
for val = w
   calc(val, A, b, TOL, x, x1);
end


function [] = calc(w, A, b, TOL, x, x1)
    n = length(b);
    for k = 1:30
        for i = 1:n
            s1 = 0;
            s2 = 0;
            for j = 1:i-1
                s1 = s1 + A(i,j)*x1(j);
            end
            for j = i+1:n
                s2 = s2 + A(i,j)*x(j);
            end
            x1(i) = (1-w)*x(i) + w*(b(i) - s1 - s2)/A(i, i);



        end
        if (norm(b - A*x1, 2)/ norm(b, 2)) < TOL
            fprintf("Resultant vector: %.6f, %.6f, %.6f, %.6f, %.6f, %.6f at iteration %d\n", x1(1), x1(2), x1(3), x1(4), x1(5), x1(6), k);
            break;
        end
        x =  x1;
    end
    if k== 30
        fprintf("Result not found\n");
    end
end
```
