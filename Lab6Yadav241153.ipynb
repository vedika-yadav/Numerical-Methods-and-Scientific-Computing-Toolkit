# <span style="color:rgb(213,80,0)">Solution to question 1</span>

```matlab
% Set the initial variables as given in the question
fprintf("Solution to Question 1 : \n");
a = 0;
b = 1;
TOL = 10^(-8);
f = @(x) x - 2^(-x);
max_iter = 30;
[p, iter] = bisection_method(f, a, b, TOL, max_iter);
fprintf("Tolerance: %e; Approximation: %.8f, Iterations: %d\n", TOL, p, iter);


function [p,iter] = bisection_method(f,a,b,TOL,max_iter)
FA = f(a);
FB = f(b);
if FA*FB > 0
    % solution can only exist in the given interval if F(a) and F(b) are of
    % different signs
    error("f(a) and f(b) must have opposite signs");
end


for iter = 1:max_iter
    % for every iteration find value of approximation
    p = a + (b-a)/2;
    FP = f(p);


    if FP == 0 || (b-a)/2 < TOL
        % if we get F(p) is zero then p is the exact correct value needed
        return
    end


    if FA*FP > 0
        % if F(p) is not zero then update conditions for the next iteration
        % if F(a) and F(p) are of different signs it means solution is in
        % (p,b)  otherwise it is in (a,p)
        a = p;
        FA = FP;
    else
        b = p;
    end


end


fprintf("Method failed after %d iterations", max_iter);
fprintf("At 30th iteration the result is - \n");
end
```


## Solution to Question 3

```matlab
fprintf("\nSolution to Question 3: \n");
% Plot y = x and y = 2sin(x)
% We'll take from 0 to 5 because solution is between 1 and 2 so it comes in
% this range and sin curve is clearly identifiable


x = linspace(0,5,1000);
y1 = x;
y2 = 2*sin(x);
plot(x,y1,'b','LineWidth',2)
hold on
plot(x,y2,'r','LineWidth',2)
xlabel('x')
ylabel('y')
title('Intersection of y = x and y = 2sin(x)')
legend('y = x','y = 2sin(x)')
grid on


% Bisection
% Initial conditions
% Solution is between 1 and 2 so we take the range from 1 to 2.
a = 1;
b = 2;
TOL = 10^(-8);
f = @(x) x - 2*sin(x);
max_iter = 30;
[p, iter] = bisection_method_3(f, a, b, TOL, max_iter);
fprintf("Tolerance: %e, Approximation = %.8f, Iteration: %d\n", TOL, p, iter);


% functions is exactly same as the first two questions
function [p,iter] = bisection_method_3(f, a, b, TOL, max_iter)
FA = f(a);
FB = f(b);
if FA*FB > 0
    % solution can only exist in the given interval if F(a) and F(b) are of
    % different signs
    error("f(a) and f(b) must have opposite signs\n");
end
p_prev = a;
for iter = 1:max_iter
    % for every iteration find value of approximation
    p = a + (b-a)/2;
    FP = f(p);


    if FP == 0 || (iter>1 && abs((p-p_prev)/p) < TOL)
        % if we get F(p) is zero then p is the exact correct value needed
        return
    end


    if FA*FP > 0
        % if F(p) is not zero then update conditions for the next iteration
        % if F(a) and F(p) are of different signs it means solution is in
        % (p,b)  otherwise it is in (a,p)
        a = p;
        FA = FP;
    else
        b = p;
    end
    p_prev = p;


end


fprintf("Method failed after %d iterations \n", max_iter);
fprintf("At 30th iteration the result is - \n");
end
```

```matlab
fprintf("\nSolution to Question 4 - \n")


f1 = @(x) (20*x + 21/(x^2))/21;
f2 = @(x) x - ((x^3) - 21)/(3*(x^2));
f3 = @(x) x - ((x^4) - 21*x)/((x^3) - 21);
f4 = @(x) (21/x)^0.5;
x0 = 1;
TOL = 10^(-10);
max_iter = 150;


[t1, p1, iter1] = fixed_point(f1, x0, TOL, max_iter);
[t2, p2, iter2] = fixed_point(f2, x0, TOL, max_iter);
[t3, p3, iter3] = fixed_point(f3, x0, TOL, max_iter);
[t4, p4, iter4] = fixed_point(f4, x0, TOL, max_iter);
fprintf("Method (a): Tolerance: %e, Approximation = %.10f, Iteration = %d, Time: %f\n", TOL, p1, iter1, t1);
fprintf("Method (b): Tolerance: %e, Approximation = %.10f, Iteration = %d, Time: %f\n", TOL, p2, iter2, t2);
fprintf("Method (c): Tolerance: %e, Approximation = %.10f, Iteration = %d, Time: %f\n", TOL, p3, iter3, t3);
fprintf("Method (d): Tolerance: %e, Approximation = %.10f, Iteration = %d, Time: %f\n", TOL, p4, iter4, t4);


fprintf("Method are ranked in the order - \n (b) > (d) > (a) > (c) \nMethod (c) is the worst because " + ...
    "it converges to 0 ie. converges to the wrong value if we start at x0 = 1. Method (b) is the fastest.\n");


function[time, p, iter] = fixed_point(f, x0, TOL, max_iter)
iter = 1;
tic;
while iter <= max_iter
    x = f(x0);
    if (abs(x - x0) < TOL)
        time = toc;
        p = x;
        return;
    end
    iter = iter+1;
    x0 = x;
end
fprintf("The method failed after %d iterations\n", max_iter);
p = x;
time = toc;


end
```
