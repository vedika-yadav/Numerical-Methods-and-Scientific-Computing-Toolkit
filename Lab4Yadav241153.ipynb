```matlab
A = [2,0,1,-1; 6,3,2,-1; 4,3,-2,3; -2,-6,2,-14]; % Matrix for given set of equations
b = [6,15,3,12]; % b vector given in quesion


x = Gaussian_elimination_partial_pivoting(A, b)




function x = Gaussian_elimination_partial_pivoting(A, b)
if isrow(b)
    b = b.'; % Ensuring b is column
end


[n,m] = size(A);
x = zeros(n, 1);  % Assuming n = m since question clearly state that solution exists


for cols = 1:n-1
    p = cols;
    maxval = abs(A(cols,cols)); % Setting initial maximum value for pivot

    for i = cols+1:n
        if abs(A(i,cols)) > maxval
            maxval = abs(A(i,cols)); % Replacing pivot value with maximum value below diagonal element
            p = i;
        end
    end


    if p~= cols
        fprintf('Step %d: swap rows %d and %d\n', cols, cols, p);
        A([cols p], :) = A([p cols], :); % Swapping the diagonal row with pivot row
        b([cols p]) = b([p cols]); % Swapping corresponding values in b
    else
        fprintf('No swap\n')
    end


    for i = cols+1:n
        mult = A(i, cols)/A(cols, cols);
        A(i, cols:m) = A(i, cols:m) - mult*A(cols, cols:m); % Forming upper diagonal matrix
        b(i) = b(i) - mult*b(cols); % Changing b correspondingly
    end



end
for i = n:-1:1
    x(i) = (b(i) - A(i, i+1:n) * x(i+1:n)) / A(i, i); % Back substitution to get solution vector
end


end
```

```matlab
A = [pi,-exp(1),sqrt(2),-sqrt(3); pi,exp(1),-exp(2),3/7; sqrt(5),-sqrt(6),1,-sqrt(2); pi^3,exp(2),-sqrt(7),1/9]; % Matrix for given set of equations
b = [sqrt(11),0,pi,sqrt(2)]; % b vector given in quesion


x = Gaussian_elimination_scaled_partial_pivoting(A, b)


function x = Gaussian_elimination_scaled_partial_pivoting(A, b)
if isrow(b)
    b = b.'; % Ensuring b is column
end


[n,m] = size(A);
x = zeros(n, 1);  % Assuming n = m since question clearly state that solution exists


s = zeros(n,1);
for i = 1:n
    s(i) = abs(A(i,1));
    for j = 2:m
        if abs(A(i,j)) > s(i)
            s(i) = abs(A(i,j));
        end
    end
end


for k = 1:n-1
    p = k;
    maxval = abs(A(k,k))/s(k); % Initialize maxval for the current row
    for i = k+1:n

        if abs(A(i,k))/s(i) > maxval
            maxval = abs(A(i,k))/s(i); % Replacing pivot value with maximum value below diagonal element for each row
            p = i;
        end
    end


    if p~= k
        fprintf('Step %d: swap rows %d and %d\n', k, k, p);
        A([k p], :) = A([p k], :); % Swapping the diagonal row with pivot row
        b([k p]) = b([p k]); % Swapping corresponding values in b
        s([k p]) = s([p k]); % swap corresponding maximum elements also
    else
        fprintf('No swap\n');
    end


    for i = k+1:n
        mult = A(i, k)/A(k, k);
        A(i, k:m) = A(i, k:m) - mult*A(k, k:m); % Forming upper diagonal matrix
        b(i) = b(i) - mult*b(k); % Changing b correspondingly
    end



end
for i = n:-1:1
    x(i) = (b(i) - A(i, i+1:n) * x(i+1:n)) / A(i, i); % Back substitution to get solution vector
end


end
```
