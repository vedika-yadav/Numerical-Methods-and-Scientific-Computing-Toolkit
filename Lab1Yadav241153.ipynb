```matlab
beta = 2;
t = 3;
L = -1;
U = 2;


vals = [];


d1 = 1;                       % first digit is always 1 (normalization)
d2_values = [0 1];            % second binary digit
d3_values = [0 1];            % third binary digit


% Step 1: construct mantissas .100, .101, .110, .111
for d2 = d2_values
    for d3 = d3_values


        % Step 2: convert binary fraction to decimal
        mantissa = d1/beta + d2/beta^2 + d3/beta^3;


        % Step 3: multiply by 2^e
        for e = L:U
            x = mantissa * beta^e;
            vals(end+1) = x;
        end
    end
end


% Step 4: include negative numbers and zero
vals = unique([0, vals, -vals]);
vals = sort(vals);


disp(vals.')


positive = vals(vals > 0);


UFL = min(positive);   % Underflow level
OFL = max(positive);   % Overflow level
N   = length(vals);    % Total elements


fprintf('UFL = %.4f\n', UFL);
fprintf('OFL = %.4f\n', OFL);
fprintf('Total elements = %d\n', N);


figure
plot(vals, zeros(size(vals)), 'o')
yticks([])
xlabel('Real line')
title('Representable numbers in F(2,3,-1,2)')
grid on
```
